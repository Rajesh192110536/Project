```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score

# Load your dataset (assuming it's in a CSV file)
data = pd.read_csv("/content/brain storm.csv")

# Data preprocessing
label_encoder = LabelEncoder()
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

numerical_cols = ['age', 'hypertension', 'avg_glucose_level', 'bmi']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Split the data into features and target
X = data.drop(columns=['stroke'])  # Features
y = data['stroke']  # Target variable

# Initialize lists to store accuracy scores
decision_tree_accuracies = []
naive_bayes_accuracies = []

# Perform 20 iterations
for i in range(20):
    # Split the data into a training and testing set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Train a Decision Tree Classifier
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)
    y_pred_dt = decision_tree.predict(X_test)
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    decision_tree_accuracies.append(accuracy_dt)

    # Train a Naive Bayes Classifier
    naive_bayes = GaussianNB()
    naive_bayes.fit(X_train, y_train)
    y_pred_nb = naive_bayes.predict(X_test)
    accuracy_nb = accuracy_score(y_test, y_pred_nb)
    naive_bayes_accuracies.append(accuracy_nb)

    # Print accuracy for each iteration
    print(f"Iteration {i+1}: Decision Tree Accuracy: {accuracy_dt}, Naive Bayes Accuracy: {accuracy_nb}")

# Calculate and print average accuracies after 30 iterations
avg_accuracy_dt = sum(decision_tree_accuracies) / len(decision_tree_accuracies)
avg_accuracy_nb = sum(naive_bayes_accuracies) / len(naive_bayes_accuracies)

print("Average Decision Tree Accuracy over 20 iterations:", avg_accuracy_dt)
print("Average Naive Bayes Accuracy over 20 iterations:", avg_accuracy_nb)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score

# Load your dataset (assuming it's in a CSV file)
data = pd.read_csv("/content/brain storm.csv")

# Data preprocessing
label_encoder = LabelEncoder()
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

numerical_cols = ['age', 'hypertension', 'avg_glucose_level', 'bmi']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Convert 'stroke' column into binary classes (0 or 1) based on a threshold (e.g., stroke > 0 indicates 1, otherwise 0)
data['stroke'] = (data['stroke'] > 0).astype(int)

# Split the data into features and target
X = data.drop(columns=['stroke'])  # Features
y = data['stroke']  # Target variable

# Initialize lists to store accuracy scores
decision_tree_accuracies = []
knn_accuracies = []

# Perform iterations
for i in range(20):
    # Split the data into a training and testing set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    # Train a Decision Tree Classifier
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)
    y_pred_dt = decision_tree.predict(X_test)
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    decision_tree_accuracies.append(accuracy_dt)

    # Using K-Nearest Neighbors Classifier
    knn_classifier = KNeighborsClassifier()
    knn_classifier.fit(X_train, y_train)
    knn_predictions = knn_classifier.predict(X_test)
    knn_accuracy = accuracy_score(y_test, knn_predictions)
    knn_accuracies.append(knn_accuracy)

    # Check if KNN accuracy is less than Decision Tree accuracy
    if knn_accuracy < accuracy_dt:
        print(f"Iteration {i+1}: KNN Accuracy ({knn_accuracy}) is less than Decision Tree Accuracy ({accuracy_dt})")

# Calculate and print average accuracies after the iterations
avg_accuracy_dt = sum(decision_tree_accuracies) / len(decision_tree_accuracies)
avg_accuracy_knn = sum(knn_accuracies) / len(knn_accuracies)

print("Average Decision Tree Accuracy over iterations:", avg_accuracy_dt)
print("Average KNN Accuracy over iterations:", avg_accuracy_knn)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import accuracy_score

# Load your dataset (assuming it's in a CSV file)
data = pd.read_csv("/content/brain storm.csv")

# Data preprocessing
# Encode categorical variables
label_encoder = LabelEncoder()
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

# Normalize numerical features
numerical_cols = ['age', 'hypertension', 'avg_glucose_level', 'bmi']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Split the data into features and target
X = data.drop(columns=['stroke'])  # Features
y = data['stroke']  # Target variable

# Initialize lists to store accuracy scores
decision_tree_accuracies = []
logistic_regression_accuracies = []

# Perform 30 iterations
for i in range(20):
    # Split the data into a training and testing set
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.01, random_state=i)

    # Train a Decision Tree Classifier
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)

    # Train a Logistic Regression Classifier
    logistic_regression = LogisticRegression()
    logistic_regression.fit(X_train, y_train)

    # Make predictions
    y_pred_dt = decision_tree.predict(X_test)
    y_pred_lr = logistic_regression.predict(X_test)

    # Evaluate the models and store accuracies
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    accuracy_lr = accuracy_score(y_test, y_pred_lr)

    decision_tree_accuracies.append(accuracy_dt)
    logistic_regression_accuracies.append(accuracy_lr)

    # Print accuracy for each iteration
    print(f"Iteration {i+1}: Decision Tree Accuracy: {accuracy_lr}, Logistic Regression Accuracy: {accuracy_dt}")

# Calculate and print average accuracies after 30 iterations
avg_accuracy_dt = sum(decision_tree_accuracies) / len(decision_tree_accuracies)
avg_accuracy_lr = sum(logistic_regression_accuracies) / len(logistic_regression_accuracies)


print("\nAverage Decision Tree Accuracy over 20 iterations:", avg_accuracy_lr)
print("Average Logistic Regression Accuracy over 20 iterations:", avg_accuracy_dt)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load your dataset (assuming it's in a CSV file)
data = pd.read_csv("/content/brain storm.csv")

# Data preprocessing
# Encode categorical variables
label_encoder = LabelEncoder()
categorical_cols = ['gender', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']
for col in categorical_cols:
    data[col] = label_encoder.fit_transform(data[col])

# Normalize numerical features
numerical_cols = ['age', 'hypertension', 'avg_glucose_level', 'bmi']
scaler = StandardScaler()
data[numerical_cols] = scaler.fit_transform(data[numerical_cols])

# Split the data into features and target
X = data.drop(columns=['stroke'])  # Features
y = data['stroke']  # Target variable

# Initialize lists to store accuracy scores
decision_tree_accuracies = []
random_forest_accuracies = []

# 20 iterations
for i in range(20):
    # Split the data into a training and testing set for each iteration
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.02, random_state=i)

    # Train a Decision Tree Classifier
    decision_tree = DecisionTreeClassifier()
    decision_tree.fit(X_train, y_train)
    y_pred_dt = decision_tree.predict(X_test)
    accuracy_dt = accuracy_score(y_test, y_pred_dt)
    decision_tree_accuracies.append(accuracy_dt)

    # Train a Random Forest Classifier
    random_forest = RandomForestClassifier(n_estimators=100)
    random_forest.fit(X_train, y_train)
    y_pred_rf = random_forest.predict(X_test)
    accuracy_rf = accuracy_score(y_test, y_pred_rf)
    random_forest_accuracies.append(accuracy_rf)

    print(f"Iteration {i+1}: Decision Tree Accuracy: {accuracy_lr}, Random Forest Accuracy: {accuracy_dt}")

# Calculate and print average accuracies after the iterations
avg_accuracy_dt = sum(decision_tree_accuracies) / len(decision_tree_accuracies)
avg_accuracy_rf = sum(random_forest_accuracies) / len(random_forest_accuracies)


print("\nAverage Decision Tree Accuracy over 20 iterations:", avg_accuracy_rf)
print("Average Random Forest Accuracy over 20 iterations:", avg_accuracy_dt)
```
